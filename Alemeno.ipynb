```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    """
    Extracts text from a given PDF file.

    Args:
        pdf_path (str): Path to the PDF file.

    Returns:
        str: Extracted text from the PDF.
    """
    text = ""
    with fitz.open(pdf_path) as pdf:
        for page_num in range(len(pdf)):
            page = pdf[page_num]
            text += page.get_text()
    return text

# Path to your uploaded file
pdf_path = "uber-10-k-2023.pdf"  # Replace with your file name
text_content = extract_text_from_pdf(pdf_path)

# Display the first 500 characters of extracted text
print(text_content[:500])
```

```python
from sentence_transformers import SentenceTransformer

def generate_embeddings(text, model_name="all-MiniLM-L6-v2"):
    """
    Generates embeddings for the given text using SentenceTransformers.

    Args:
        text (str): Input text for embedding generation.
        model_name (str): Pre-trained model to use for embeddings.

    Returns:
        list: Sentences and their corresponding embeddings.
    """
    model = SentenceTransformer(model_name)
    sentences = text.split('\n')  # Split text into smaller chunks (by lines)
    embeddings = model.encode(sentences, convert_to_tensor=True)
    return sentences, embeddings

# Generate embeddings for the extracted text
sentences, embeddings = generate_embeddings(text_content)

# Check the number of embeddings generated
print(f"Generated {len(embeddings)} embeddings for the document.")
```

```python
import chromadb
from chromadb.config import Settings

# Initialize ChromaDB client
client = chromadb.Client(Settings(persist_directory="./vector_db"))

# Create a new collection (or access an existing one)
collection = client.get_or_create_collection("pdf_embeddings")

# Add sentences and their embeddings to the collection
for sentence, embedding in zip(sentences, embeddings):
    collection.add(
        documents=[sentence],
        metadatas=[{"sentence": sentence}],
        embeddings=[embedding]
    )

print("Embeddings stored successfully!")
```
